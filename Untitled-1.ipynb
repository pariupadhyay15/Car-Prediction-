```python
car=car[car['year'].str.isnumeric()]
#filters out rows that contain numeric values in year column
```

```python
car['year']=car['year'].astype(int)
```

```python
car= car[car["Price"]!="Ask For Price"]
#removes rows which conatin ask for price in price column
```

```python
car['Price']=car['Price'].str.replace(",","").astype(int)
#replaces comma by empty string in price column
```

```python
car['kms_driven']=car['kms_driven'].str.split(" ").str.get(0).str.replace(",","")
# the kms driven col has km writeen in it so by this the first str before the space is taken in consideration 
#commas are replaced by string in the same cell
```

```python
car=car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven']=car['kms_driven'].astype(int)
```

```python
car=car[~car['fuel_type'].isna()]
#keeps col of fuel type which are not null
```

```python
car['name']=car['name'].str.split(" ").str.slice(0,3).str.join(' ')
#the name of car is very big so we are keeping only the first 3 words
```

```python
car.reset_index(drop=True)
#some of the indices got droped beacuse of the preprocesing so new indices are given here
```

```python
car.info()
```

```python
car.describe()
```

```python
car=car[car['Price']<6e6].reset_index(drop=True)
#only 1 car is there which has greater value than 60lac so we are going to remove it
```

```python
car.to_csv("cleaned car.csv")
```

```python
X=car.drop(columns="Price")
y= car['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X,y, test_size=0.2, random_state=661)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe= OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_), ['name','company','fuel_type']),
                                     remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
#by this pipeline we can send in the X_train, y_train to tranform the cols
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python
r2_score(y_test, y_pred)
```
